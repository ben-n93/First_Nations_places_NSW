```python
import pandas as pd
import geojson
#import folium
```

```python
df = pd.read_csv('GNR_27Jul2023.csv',
    dtype={
    'REFERENCE': str,
    'PLACENAME': str,
    'DESIGNATION': str,
    'STATUS': str,
    'DUAL NAME': str,
    'GAZETTE DATE': str,
    'GEOGRAPHICAL NAME': str,
    'PREVIOUS NAMES': str,
    'GNB FILE': str,
    'LGA': str,
    'LGA AT GAZETTAL': str,
    '1:25,000 MAP NAME': str,
    '1:100,000 MAP': str,
    'PARISH': str,
    'COUNTY': str,
    'DESCRIPTION': str,
    'MEANING': str,
    'ORIGIN': str,
    'HISTORY': str,
    'PRONUNCIATION': str,
    'ABORIGINAL NAME': str,
    'GDA2020 LAT': float })

pd.set_option('display.max_columns', None)
df.head()
```

```python
aboriginal_places = df.loc[df['ABORIGINAL NAME'] == 'Yes']
aboriginal_places.head()
```

```python
meaning_places = df[df['MEANING'].str.contains('aboriginal', na=False)]
meaning_places.head()
```

```python
origin_places = df[df['ORIGIN'].str.contains('aboriginal', na=False)]
origin_places.head()
```

```python
first_nations_named_places = pd.concat([aboriginal_places, meaning_places, origin_places])
first_nations_named_places = first_nations_named_places.drop_duplicates()
first_nations_named_places.head()
```

```python
# Remove unwanted columns and rows with null coordinates.
first_nations_named_places = first_nations_named_places[['PLACENAME', 'DESIGNATION', 'STATUS', 'DUAL NAME', 'GEOGRAPHICAL NAME', 'GDA2020 LAT', 'GDA2020 LONG']]
first_nations_named_places.dropna(subset=['GDA2020 LAT'], inplace=True)
pd.isna(df.at[2, 'A']).dropna(subset=['GDA2020 LONG'], inplace=True)
```

```python
# Create GeoJSON file.
features = []
for index, row in first_nations_named_places.iterrows():
    coordinates = geojson.Point((row['GDA2020 LONG'], row['GDA2020 LAT']))
    properties = first_nations_named_places.loc[index, ['PLACENAME', 'DESIGNATION', 'STATUS', 
                                                        'DUAL NAME', 'GEOGRAPHICAL NAME']].to_dict()
    # Nan values (Numpy nan object) returned by Pandas for null values 
    # will result in error when passed to GeoJSON.
    new_properties = {}
    for key, value in properties.items():
        if pd.isna(first_nations_named_places.at[index, key]):
            new_properties[key] = None
        else:
            new_properties[key] = value
        
    # Creature feature.
    feature = geojson.Feature(geometry=coordinates, properties=new_properties)
    features.append(feature)
```
